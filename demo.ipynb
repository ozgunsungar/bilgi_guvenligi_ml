# Data Preprocessing

```python
df_clean = df.copy()
duplicate_rows = df_clean[df_clean.duplicated()]
duplicated_counts = duplicate_rows.shape[0]

if duplicated_counts > 0 :
    #we have duplicated rows
    df_clean.drop_duplicates(inplace=True)
    print('number of duplicates dropped --> {}'.format(duplicated_counts))
else : 
    print("there are no duplicated rows to get deleted")
```

```python
from imblearn.over_sampling import SMOTE

#Fixing the imbalance using SMOTE Technique
df_smote = df_clean.copy()
print("Class distribution before SMOTE")
print(df_clean['spam'].value_counts())

columns = df_clean.columns
X = df_clean.iloc[:,:-1]
y = df_clean.iloc[:,-1]

smote=SMOTE()
X,y = smote.fit_resample(X,y)
df_smote = pd.DataFrame(X,columns=columns)
df_smote['spam']=y

print('\nClass distribution after applying SMOTE Technique:',)
print(y.value_counts())
```

# Data Manipulation

```python
df_smote.columns
```

```python
df_manipulated = df_smote.copy()
df_manipulated.columns=[i.replace('[','_') for i in df_manipulated.columns]

X = df_manipulated.iloc[:,:-1]
y = df_manipulated.iloc[:,-1]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, 
                                                    test_size=0.2, random_state=0)

print('Original set  ---> ',X.shape,y.shape,
      '\nTraining set  ---> ',X_train.shape,y_train.shape,
      '\nTesting set   ---> ', X_test.shape,'', y_test.shape)
```

```python
# Feature Scaling (Standardization)

from sklearn.preprocessing import StandardScaler
std = StandardScaler()

print('Standardardization on Training set'.center(100))
X_train_std = std.fit_transform(X_train)
X_train_std = pd.DataFrame(X_train_std, columns=X.columns)
display(X_train_std.describe())

print('\nStandardardization on Testing set'.center(100))
X_test_std = std.transform(X_test)
X_test_std = pd.DataFrame(X_test_std, columns=X.columns)
display(X_test_std.describe())
```

# PCA

```python
from sklearn.decomposition import PCA

pca = PCA().fit(X_train_std)

fig, ax = plt.subplots(figsize=(14,6))
x_values = range(1, pca.n_components_+1)
ax.bar(x_values, pca.explained_variance_ratio_, lw=2, label='Explained Variance')
ax.plot(x_values, np.cumsum(pca.explained_variance_ratio_), lw=2, label='Cumulative Explained Variance', color='red')
plt.plot([0,pca.n_components_+1],[0.90,0.90],'g--')
plt.plot([43,43],[0,1], 'g--')
ax.set_title('Explained variance of components')
ax.set_xlabel('Principal Component')
ax.set_ylabel('Explained Variance')
plt.grid()
plt.legend()
plt.show()
```

```python
#Let us create first create a table to store the results of various models 

Evaluation_Results = pd.DataFrame(np.zeros((4,5)), 
                                  columns=['Accuracy', 'Precision','Recall','F1-score','AUC-ROC score'])

Evaluation_Results.index=['Support Vector Machine (SVM)', 'Cat-Boost',
                          'Light Gradient Boosting Machine (Light GBM)',
                         'Artificial Neural Network (ANN)']
Evaluation_Results
```

```python
#Let us define functions to summarise the Prediction's scores .
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,confusion_matrix,classification_report
import scikitplot as skplt

#Classification Summary Function
def Classification_Summary(y_true,pred,pred_prob,i):
    Evaluation_Results.iloc[i]['Accuracy']=round(accuracy_score(y_test, pred),3)*100   
    Evaluation_Results.iloc[i]['Precision']=round(precision_score(y_test, pred, average='weighted'),3)*100 #
    Evaluation_Results.iloc[i]['Recall']=round(recall_score(y_test, pred, average='weighted'),3)*100 #
    Evaluation_Results.iloc[i]['F1-score']=round(f1_score(y_test, pred, average='weighted'),3)*100 #
    Evaluation_Results.iloc[i]['AUC-ROC score']=round(roc_auc_score(y_test, pred_prob[:,1], multi_class='ovr'),3)*100 #[:, 1]
    print('{}{}\033[1m Evaluating {} \033[0m{}{}\n'.format('<'*3,'-'*25,Evaluation_Results.index[i], '-'*35,'>'*25))
    print('Accuracy = {}%'.format(round(accuracy_score(y_test, pred),3)*100))
    print('F1 Score = {}%'.format(round(f1_score(y_test, pred, average='weighted'),3)*100)) #
    print('\n \033[1mConfusiton Matrix:\033[0m\n',confusion_matrix(y_test, pred))
    print('\n\033[1mClassification Report:\033[0m\n',classification_report(y_test, pred))
    
    skplt.metrics.plot_roc_curve(y_test, pred_prob)
    plt.show()
```

# SVM

```python
# Building Support Vector Machine Classifier
from sklearn import svm
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import RandomizedSearchCV

SVM_model = svm.SVC(probability=True).fit(X_train_std, y_train)

svm_param = {"C": [.01, .1, 1, 5, 10, 100],             
             "gamma": [.01, .1, 1, 5, 10, 100],
             "kernel": ["rbf"],
             "random_state": [1]}

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

RCV = RandomizedSearchCV(SVM_model, svm_param, n_iter=50, scoring='roc_auc', n_jobs=-1, cv=5, random_state=1)

SVM = RCV.fit(X_train_std, y_train).best_estimator_
pred = SVM.predict(X_test_std)
pred_prob = SVM.predict_proba(X_test_std)
```

```python
Classification_Summary(y_test,pred,pred_prob,0)
```

# LightGBM

```python
# Building Support Vector Machine Classifier
import lightgbm as lgb

lgb_model = lgb.LGBMClassifier().fit(X_train_std, y_train)

lgb_param = {'bagging_fraction': (0.5, 0.8),
            'bagging_frequency': (5, 8),
            'feature_fraction': (0.5, 0.8),
            'max_depth': (10, 13),
            'min_data_in_leaf': (90, 120),
            'num_leaves': (1200, 1550)}

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

RCV = RandomizedSearchCV(lgb_model, lgb_param, n_iter=50, scoring='roc_auc', n_jobs=-1, cv=5, random_state=1)

LGB = RCV.fit(X_train_std, y_train).best_estimator_
pred = LGB.predict(X_test_std)
pred_prob = LGB.predict_proba(X_test_std)
```

```python
Classification_Summary(y_test,pred,pred_prob,2)
```
